```python
import smtplib
from email.mime.text import MIMEText
from langchain_ollama import OllamaLLM
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
import PyPDF2

# ----------------------
# Step 1: Extract Text from PDF
# ----------------------
def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            text += page.extract_text() + "\n"
    return text

# ----------------------
# Step 2: Summarize with Ollama
# ----------------------
def summarize_text(text):
    prompt_template = PromptTemplate(
        input_variables=["content"],
        template="Summarize the following text in clear bullet points:\n\n{content}"
    )

    llm = OllamaLLM(model="llama2")   # or "mistral", "gemma", etc.
    chain = LLMChain(llm=llm, prompt=prompt_template)
    summary = chain.run(content=text)
    return summary

# ----------------------
# Step 3: Send Email
# ----------------------
def send_email(subject, body, sender, recipient, password, smtp_server="smtp.gmail.com", port=587):
    msg = MIMEText(body)
    msg["Subject"] = subject
    msg["From"] = sender
    msg["To"] = recipient

    with smtplib.SMTP(smtp_server, port) as server:
        server.starttls()
        server.login(sender, password)
        server.sendmail(sender, recipient, msg.as_string())
        print("✅ Email sent successfully!")

# ----------------------
# Run Multi-step Agent
# ----------------------
if __name__ == "__main__":
    # 1. Extract from PDF
    pdf_path = r"C:\Users\ASUS\OneDrive\Desktop\ai.pdf"  # put your PDF file name here
    raw_text = extract_text_from_pdf(pdf_path)

    # 2. Summarize
    summary = summarize_text(raw_text)

    # 3. Email
    sender_email = "kuhukhandelwal2273@gmail.com"
    receiver_email = "kuhukhandelwal2711@gmail.com"
    app_password = "tahm earc ulvf bbhs"   # Use app password for Gmail

    send_email("PDF Summary Report", summary, sender_email, receiver_email, app_password)
```
