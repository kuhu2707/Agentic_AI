```python
from crewai import Agent, Task, Crew
from langchain_ollama import OllamaLLM

llm = OllamaLLM(model="llama3")  

researcher = Agent(
    role="Researcher",
    goal="Find detailed information about the given topic.",
    backstory="You are a skilled researcher who knows how to extract the most relevant facts.",
    llm=llm,
    verbose=True
)

writer = Agent(
    role="Writer",
    goal="Summarize the research into a short, clear report.",
    backstory="You are a professional writer who explains information in simple words.",
    llm=llm,
    verbose=True
)


research_task = Task(
    description="Research about the impact of Artificial Intelligence in education.",
    agent=researcher,
    expected_output="A list of key points about AI's impact on education."
)

writing_task = Task(
    description="Write a summary report based on the researcher's findings.",
    agent=writer,
    expected_output="A well-structured summary report in simple language."
)


crew = Crew(
    agents=[researcher, writer],
    tasks=[research_task, writing_task],
    verbose=True
)

result = crew.kickoff()
print("\n Final Report:\n", result)
```
